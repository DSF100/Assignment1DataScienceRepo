```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
print(1)
```

```python
LANGUAGE_DATA_2016_C = "2016Census_G13C_AUS.csv"
LANGUAGE_DATA_2016_D = "2016Census_G13D_AUS.csv"

LANGUAGE_DATA_2021_C = "2021Census_G13C_AUS_AUS.csv"
LANGUAGE_DATA_2021_D = "2021Census_G13D_AUS_AUS.csv"
LANGUAGE_DATA_2021_E = "2021Census_G13E_AUS_AUS.csv"
```

```python
list_data_2016 = []
list_data_2021 = []

with open(LANGUAGE_DATA_2016_C, newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    data = list(csvreader)
    for name in data[0]:
        if name[0] == "P":
            list_data_2016.append(name)

with open(LANGUAGE_DATA_2016_D, newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    data = list(csvreader)
    for name in data[0]:
        if name[0] == "P":
            list_data_2016.append(name)

with open(LANGUAGE_DATA_2021_C, newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    data = list(csvreader)
    for name in data[0]:
        if name[0] == "P":
            list_data_2021.append(name)

with open(LANGUAGE_DATA_2021_D, newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    data = list(csvreader)
    for name in data[0]:
        if name[0] == "P":
            list_data_2021.append(name)

with open(LANGUAGE_DATA_2021_E, newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    data = list(csvreader)
    for name in data[0]:
        if name[0] == "P":
            list_data_2021.append(name)

print(len(list_data_2016))
print(len(list_data_2021))
#print(list_data_2016)
#print(list_data_2021)

i = 0
for i in range(264):
    print(list_data_2016[i] + "       " + list_data_2021[i])
```
